
### ==== PASSO 2: FUNÇÕES DE BUSCA NA API  ====


```python
# Função original para buscar dados de um ISBN conhecido
def get_book_data_by_isbn(isbn):
    try:
        url = f"https://openlibrary.org/api/books?bibkeys=ISBN:{isbn}&format=json&jscmd=data"
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        data = response.json()
        return data.get(f"ISBN:{isbn}")
    except requests.exceptions.RequestException as e:
        print(f"  - ERRO (get_book_data_by_isbn): Falha na requisição para o ISBN {isbn}: {e}")
        return None
```


### ==== PASSO 3: CONSTRUINDO O DATAFRAME DINAMICAMENTE ====


```python
livros = []
isbns_processados = set()  # Para evitar livros duplicados
max_livros_por_disciplina = 20 # Limite para não ficarmos aqui para sempre e diversificar

for disciplina, subjects in subjects_por_disciplina.items():
    livros_encontrados_disciplina = 0
    print(f"\n[DISCIPLINA: {disciplina}]")

    for subject in subjects:
        if livros_encontrados_disciplina >= max_livros_por_disciplina:
            break # Passa para a próxima disciplina se já atingimos o limite

        try:
            # Busca obras populares sobre o assunto
            url_subject = f"https://openlibrary.org/subjects/{subject}.json?limit=10"
            print(f"  Buscando assunto '{subject}'...")
            response = requests.get(url_subject, timeout=15)
            response.raise_for_status()
            data = response.json()

            # Para cada obra encontrada, tenta achar um ISBN
            for work in data.get('works', []):
                if livros_encontrados_disciplina >= max_livros_por_disciplina:
                    break

                # A API de assuntos nos dá uma chave para cada obra
                work_key = work.get('key')
                if not work_key:
                    continue

                # Busca as edições daquela obra para encontrar um ISBN
                url_editions = f"https://openlibrary.org{work_key}/editions.json?limit=5"
                editions_response = requests.get(url_editions, timeout=10)
                editions_data = editions_response.json()

                # Pega o primeiro ISBN válido que encontrar e que ainda não processamos
                for edition in editions_data.get('entries', []):
                    if 'isbn_13' in edition and edition['isbn_13'][0] not in isbns_processados:
                        isbn = edition['isbn_13'][0]
                        print(f"    > ISBN encontrado: {isbn}. Buscando detalhes...")

                        book_data = get_book_data_by_isbn(isbn)

                        if book_data:
                            autores = [autor.get('name', 'N/A') for autor in book_data.get('authors', [])]
                            editoras = [editora.get('name', 'N/A') for editora in book_data.get('publishers', [])]

                            livros.append({
                                "ISBN": isbn,
                                "Título": book_data.get("title", "N/A"),
                                "Autor": ", ".join(autores) if autores else "N/A",
                                "Editora": ", ".join(editoras) if editoras else "N/A",
                                "Ano": book_data.get("publish_date", "N/A"),
                                "Páginas": book_data.get("number_of_pages", 0),
                                "Disciplina": disciplina,
                                "Capa": book_data.get("cover", {}).get("large", "")
                            })

                            print(f"    >> SUCESSO! Livro '{book_data.get('title', 'N/A')}' adicionado.")
                            isbns_processados.add(isbn)
                            livros_encontrados_disciplina += 1
                            break # Para de procurar edições para esta obra

                time.sleep(1) # Pausa de 1 segundo para ser gentil com a API

        except requests.exceptions.RequestException as e:
            print(f"  - ERRO (main_loop): Falha ao buscar o assunto '{subject}': {e}")
            continue
        except Exception as e:
            print(f"  - ERRO (main_loop): Erro inesperado no assunto '{subject}': {e}")
            continue
```


### ==== PASSO 4: FINALIZAR E SALVAR O DATAFRAME ====

```python
df_livros = pd.DataFrame(livros)

if not df_livros.empty:
    # Adicionando dados sintéticos da biblioteca
    df_livros["Qtd_Exemplares"] = np.random.randint(1, 10, size=len(df_livros))
    df_livros["Avaliação"] = np.round(np.random.uniform(3.5, 5.0, size=len(df_livros)), 1)
    df_livros["Empréstimos"] = np.random.randint(0, 100, size=len(df_livros))

    # Ordenando
    df_livros = df_livros.sort_values(by="Disciplina").reset_index(drop=True)

    # Salvando
    df_livros.to_csv("biblioteca_dinamica.csv", index=False)

    print("\n" + "="*60)
    print(f"PROCESSO CONCLUÍDO! DataFrame final criado com {len(df_livros)} livros!")
    print("Amostra dos dados:")
    print(df_livros.head())
    print("="*60)
else:
    print("\nNenhum livro foi adicionado ao DataFrame. Isso pode ocorrer por problemas de conexão ou com a API da Open Library.")
```
